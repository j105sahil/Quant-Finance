LOADING BAJAJ FINSV AS df_bajaj


```python
import csv 
import pandas as pd
with open('NSE-BAJAJFINSV.csv', 'r') as f:
    reader = csv.reader(f)
    count=0
    data_bajaj = []   
    for row in reader:
        if count==0:
            header = row
            count +=1
        else:
            data_bajaj.append(row)
            count +=1
            
df_bajaj = pd.DataFrame(data_bajaj, columns=header)
#Converting string into float
try :
    for i in range(len(df_bajaj['Open'])):
        df_bajaj['Open'][i]=float( df_bajaj['Open'][i])
    for i in range(len(df_bajaj['High'])):
        df_bajaj['High'][i]=float( df_bajaj['High'][i])
    for i in range(len(df_bajaj['Low'])):
        df_bajaj['Low'][i]=float( df_bajaj['Low'][i])
    for i in range(len(df_bajaj['Last'])):
        df_bajaj['Last'][i]=float( df_bajaj['Last'][i])
    for i in range(len(df_bajaj['Close'])):
        df_bajaj['Close'][i]=float( df_bajaj['Close'][i])
    for i in range(len(df_bajaj['Total Trade Quantity'])):
        df_bajaj['Total Trade Quantity'][i]=float( df_bajaj['Total Trade Quantity'][i])
    for i in range(len(df_bajaj['Turnover (Lacs)'])):
        df_bajaj['Turnover (Lacs)'][i]=float( df_bajaj['Turnover (Lacs)'][i])
except:
    pass
```

```python
#Calculating weight
lot_bajaj = 125
lot_ITC = 2400
lot_tech = 1200
lot_titan = 750
lot_hind = 3500
total_lot = lot_bajaj + lot_ITC + lot_tech + lot_titan + lot_hind
w_bajaj = lot_bajaj/total_lot*5000000
w_ITC = lot_ITC/total_lot*5000000
w_tech = lot_tech/total_lot*5000000
w_titan = lot_titan/total_lot*5000000
w_hind = lot_hind/total_lot*5000000
```

# Daily P&l

```python
def pnl_func(close_buy,close_sell,weight):
    no_of_stock=weight//close_buy
    buy = close_buy*no_of_stock
    sold = close_sell*no_of_stock
    pnl = sold - buy
    return pnl
```

```python
import pandas as pd
import matplotlib.pyplot as plt
Profit_Loss_bajaj = []

for i in range(1,len(df_bajaj10)):
    buying = df_bajaj10['Close'][i-1]
    selling = df_bajaj10['Close'][i]
    pnl = pnl_func(buying,selling,w_bajaj)
    Profit_Loss_bajaj.append(pnl)

Profit_Loss_ITC = []
for i in range(1,len(df_ITC10)):
    buying = df_ITC10['Close'][i-1]
    selling = df_ITC10['Close'][i]
    pnl = pnl_func(buying,selling,w_ITC)
    Profit_Loss_ITC.append(pnl)

Profit_Loss_tech = []
for i in range(1,len(df_tech10)):
    buying = df_tech10['Close'][i-1]
    selling = df_tech10['Close'][i]
    pnl = pnl_func(buying,selling,w_tech)
    Profit_Loss_tech.append(pnl)
    
Profit_Loss_titan = []
for i in range(1,len(df_titan10)):
    buying = df_titan10['Close'][i-1]
    selling = df_titan10['Close'][i]
    pnl = pnl_func(buying,selling,w_titan)
    Profit_Loss_titan.append(pnl)

Profit_Loss_hind = []
for i in range(1,len(df_hind10)):
    buying = df_hind10['Close'][i-1]
    selling = df_hind10['Close'][i]
    pnl = pnl_func(buying,selling,w_hind)
    Profit_Loss_hind.append(pnl)

daily_pnl = []
for i in range(len(Profit_Loss_bajaj)):
    X = Profit_Loss_bajaj[i] + Profit_Loss_ITC[i] + Profit_Loss_tech[i] + Profit_Loss_titan[i] + Profit_Loss_hind[i]
    daily_pnl.append(X)
df_daily_pnl = pd.DataFrame(daily_pnl)
fig = df_daily_pnl[0].plot(grid=True,figsize=(20,15))
```

# Value At Risk

```python
from scipy.stats import norm 
def value_at_risk(mean,std):
    Var_90 = norm.ppf(1-0.9,mean,std)
    return Var_90
```
